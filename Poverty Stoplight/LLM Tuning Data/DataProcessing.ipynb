```python
class Family:
    def __init__(self, id, members, income, accessToCredit,  familySavings, diversifiedSourcesOfIncome, documentation, drinkingWaterAccess, garbageDisposal, accessToHealthServices, alimentation, eyesight, dentalCare, personalHygiene, unpollutedEnvironment, insurance, safeHouse, safeBathroom, refrigerator, separateBedrooms, properKitchen, regularMeansOfTransportation, road, phone, safety, securityOfProperty, electricityAccess, clothingAndFootwear, readAndWrite, middleEducation, knowledgeAndSkillsToGenerateIncome, capacityToPlanAndBudget, schoolSuppliesAndBooks, accessInformation, accessToEntertainment, culturalTraditionsAndHeritage, respectForDiversity, awarenessOfHumanRights, influenceInPublicSector, abilityToSolveProblemsAndConflicts, registeredToVoteAndVotesInElections, socialCapital, awarenessOfNeeds, selfEsteem, moralConscience, emotionalIntelligence, entrepreneurialSpirit, autonomyDecisions):
        self.id = id
        self.members = members
        self.income = income
        self.accessToCredit = accessToCredit
        self.familySavings = familySavings
        self.diversifiedSourcesOfIncome = diversifiedSourcesOfIncome
        self.documentation = documentation
        self.drinkingWaterAccess = drinkingWaterAccess
        self.garbageDisposal = garbageDisposal
        self.accessToHealthServices = accessToHealthServices
        self.alimentation = alimentation
        self.eyesight = eyesight
        self.dentalCare = dentalCare
        self.personalHygiene = personalHygiene
        self.unpollutedEnvironment = unpollutedEnvironment
        self.insurance = insurance
        self.safeHouse = safeHouse
        self.safeBathroom = safeBathroom
        self.refrigerator = refrigerator
        self.separateBedrooms = separateBedrooms
        self.properKitchen = properKitchen
        self.regularMeansOfTransportation = regularMeansOfTransportation
        self.road = road
        self.phone = phone
        self.safety = safety
        self.securityOfProperty = securityOfProperty
        self.electricityAccess = electricityAccess
        self.clothingAndFootwear = clothingAndFootwear
        self.readAndWrite = readAndWrite
        self.middleEducation = middleEducation
        self.knowledgeAndSkillsToGenerateIncome = knowledgeAndSkillsToGenerateIncome
        self.capacityToPlanAndBudget = capacityToPlanAndBudget
        self.schoolSuppliesAndBooks = schoolSuppliesAndBooks
        self.accessInformation = accessInformation
        self.accessToEntertainment = accessToEntertainment
        self.culturalTraditionsAndHeritage = culturalTraditionsAndHeritage
        self.respectForDiversity = respectForDiversity
        self.awarenessOfHumanRights = awarenessOfHumanRights
        self.influenceInPublicSector = influenceInPublicSector
        self.abilityToSolveProblemsAndConflicts = abilityToSolveProblemsAndConflicts
        self.registeredToVoteAndVotesInElections = registeredToVoteAndVotesInElections
        self.socialCapital = socialCapital
        self.awarenessOfNeeds = awarenessOfNeeds
        self.selfEsteem = selfEsteem
        self.moralConscience = moralConscience
        self.emotionalIntelligence = emotionalIntelligence
        self.entrepreneurialSpirit = entrepreneurialSpirit
        self.autonomyDecisions = autonomyDecisions

class Person:
    def __init__(self, id: int, date: str, gender: str, birthday: str, age: int, primary : bool, education, famCode: str):
        self.id = id
        self.date = date
        self.geder = gender
        self.birthday = birthday
        self.age = age
        self.primary = primary
        self.education = education
        self.famCode = famCode
```
